```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import pylab as pl

# user Cluster
user=np.load('../../data/latent_user.npy')
print(user)
```

```python
# kmeans 클러스터링
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
score = [kmeans[i].fit(user).score(user) for i in range(len(kmeans))]
print(score)
```

```python
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

```python
kmeans=KMeans(n_clusters=7).fit(user)

df=pd.DataFrame(user)
df['cluster'] = kmeans.labels_
print(df)
```

```python
# 각 군집별 평균값
kmeans_cluster_mean=df.groupby(['cluster']).mean()
print(kmeans_cluster_mean)
```

```python
# movie Cluster
movie=np.load('../../data/latent_movie.npy')
print(movie)
```

```python
# kmeans 클러스터링
Nc = range(1, 20)
movie_kmeans = [KMeans(n_clusters=i) for i in Nc]
movie_score = [movie_kmeans[i].fit(movie).score(movie) for i in range(len(movie_kmeans))]
print(movie_score)
```

```python
pl.plot(Nc,movie_score)
pl.xlabel('Number of Movie Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```
